```python
def find_best_len(max_len, lst):
    count = 0
    for sentence in lst:
        if(sentence <= max_len):
            count += 1
    print('전체 샘플 중 길이가 %s 이하인 샘플의 비율: %s' %(max_len, (count / len(lst))*100))
```

```python
max_len = 200
tmp_lst

find_best_len(max_len, tmp_lst)
```

```python
max_len = 300
tmp_lst

find_best_len(max_len, tmp_lst)
```
