```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
from collections import Counter
from konlpy.tag import Mecab
from sklearn.model_selection import train_test_split
import time
import csv
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from konlpy.tag import Twitter
import konlpy
from konlpy.tag import Kkma, Komoran, Hannanum, Okt
from konlpy.utils import pprint
from collections import Counter


f = open('/content/불용어_0418_2.txt', 'r')
stopword  = []
for word in f:
    word = word.replace('\n','')
    stopword.append(word)

print(stopword)
# 파일과 형태소 분석기 설정
df = pd.read_csv('/content/total_final.csv')
okt = Okt()
# mecab = Mecab()
# #hannanum = Hannanum()
# #kkma = Kkma()
list1 = []
문장 = []   
리뷰모음 = df.review
result = []
i=0
def 형태소분석후_불용어제거_워드클라우드(df):

    리뷰모음 = df.review
    for 리뷰 in 리뷰모음:

        형태소분석 = okt.morphs(리뷰,stem=True)
        #print(형태소분석)
        result = []
        for word in 형태소분석:
            
            if word not in stopword: 
                # print(word)
                result.append(word)
                list1.append(word)


            
                
            # i+= 1
            # if i <  50:
            #     break
        print(result)
        with open("total_불용어제거_0418.csv", 'a+',encoding='utf-8-sig') as file: 
            writer = csv.writer(file) 
            result = [x for x in result if len(x)>1]
            writer.writerow(result)


    # 한 글자 제거
    remove_one_word = [x for x in list1 if len(x)>1]

    words = Counter(remove_one_word)
    #word2 = Counter(remove_one_word)
    tags = words.most_common(100)

    c = dict(tags)

    # # 워드클라우드 설정 부분   - 폰트파일(ttf) 필요
    wc = WordCloud(font_path='/content/Hancom Gothic Bold.ttf',width=400, height=400, scale=2.0, max_font_size=250)
    cloud = wc.generate_from_frequencies(c)

    plt.figure(figsize=(10, 8))
    plt.axis('off')
    plt.imshow(cloud)
    plt.show()
    plt.savefig('워드클라우드_mecab.png',bbox_inches='tight')

    return print('완료')

형태소분석후_불용어제거_워드클라우드(df)
```
